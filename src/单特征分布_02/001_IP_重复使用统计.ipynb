```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. 读取数据

```python
# 读取csv 文件
df = pd.read_csv('../../data/周一所获数据/所有IP的学员使用数量.csv')

# 查看数据
df.describe()
```

# 2. 统计IP重复使用次数
因为源数据已经有序,因此没有额外做排序处理


```python
ip_count = df.iloc[:, [0, 2]]
ip_count.head(10)
ip_count.dtypes
```

# 统计IP地址重复使用次数

**直方图形式**

```python
# 要求图表中只有 1 行 1 列子图，并且设置图表的大小为 10 * 10
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 20))

# 绘制柱状图
num_elements = 40
ax.bar(ip_count.head(num_elements)._id, ip_count.head(num_elements).uniqueTrainingPersonIdsCount)

# 设置标题
ax.set_xlabel('IP address')
ax.set_ylabel('Count')

# 设置x轴标签倾斜45度
plt.xticks(rotation=45)

# 设置 y 轴刻度精度
plt.yticks(np.arange(0, 3000, 100))

plt.savefig('../../imgs/所有IP的学员使用数量1-6.png', dpi=800, bbox_inches='tight', pad_inches=1)
plt.show()
```
