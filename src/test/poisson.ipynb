```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# 设置中文显示
from pylab import mpl

mpl.rcParams['font.sans-serif'] = ['Microsoft YaHei']  # 指定默认字体：解决plot不能显示中文问题
mpl.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
```

```python
 # λ = 实验次数n × 概率P
mu = 8  # 泊松分布的期望值（均值）

# 事件发生的次数
x = np.arange(0, 100)

# 计算泊松分布的概率质量函数
pmf = poisson.pmf(x, mu)
```

```python

# 计算累积分布函数
cdf = poisson.cdf(x, mu)
```

具体绘制图形：https://www.runoob.com/matplotlib/matplotlib-marker.html

```python
# 具体绘制图形：https://www.runoob.com/matplotlib/matplotlib-marker.html
# 绘制概率质量函数
plt.figure(figsize=(10, 5))
# plt.stem(x, pmf, basefmt='C0--', use_line_collection=True,) # 棉棒图
plt.plot(x, pmf, 'C0-', label='PMF')
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
# plt.xticks(np.arange(0, x.max(), 5))
plt.yticks()
plt.show()
```

```python
plt.rcParams['axes.linewidth'] = 6
fig = plt.figure(figsize=(16, 8))
plt.title("泊松分布")

# --子图1--
ax1 = fig.add_subplot(221)

ax1.set_xlabel('事件发生的次数(λ=8)')
ax1.set_ylabel('概率')
ax1.plot(x, poisson.pmf(x, 8), 'C0-', label='PMF')


# --子图2--
ax2 = fig.add_subplot(222)

ax2.set_xlabel('事件发生的次数(λ=25)')
ax2.set_ylabel('概率')
ax2.plot(x, poisson.pmf(x, 25), 'C0-', label='PMF')


# --子图3--
ax3 = fig.add_subplot(223)

ax3.set_xlabel('事件发生的次数(λ=50)')
ax3.set_ylabel('概率')
ax3.plot(x, poisson.pmf(x, 50), 'C0-', label='PMF')


# --子图4--
ax4 = fig.add_subplot(224)

ax4.set_xlabel('事件发生的次数(λ=80)')
ax4.set_ylabel('概率')
ax4.plot(x, poisson.pmf(x, 80), 'C0-', label='PMF')

# 使用 tight_layout 自动调整间距
fig.tight_layout()

# plt.savefig("./imgs/MeanShift-3-features-with-SuspiciousUser(归一化).png")
fig.show()
```
