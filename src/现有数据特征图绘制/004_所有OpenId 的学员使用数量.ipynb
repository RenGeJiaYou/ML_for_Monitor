```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. 读取数据

```python
# 读取csv 文件并指定数据类型呢
file_path = '../../data/周一所获数据/所有OpenId的学员使用数量.csv'
df = pd.read_csv(file_path, dtype={'uniqueStudentIdCount': 'int64'})

# 查看数据
df = df.iloc[1:]
df.head()
```

# 2. 绘制柱状图

```python
# 要求图表中只有 1 行 1 列子图，并且设置图表的大小为 10 * 10
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 20))

num_elements = 5000
total_rows = df.shape[0]

# x 轴不需要打印每一条 id，只需要打印刻度
x = np.arange(total_rows)
ax.bar(x, df.head(total_rows).uniqueStudentIdCount)

# 设置标题
ax.set_xlabel('OpenID')
ax.set_ylabel('uniqueStudentIdCount')

# 设置 x 轴刻度
ticks = np.arange(0, total_rows, 1000)
labels = [f"{t}" for t in ticks]
plt.xticks(ticks, labels)

# 设置 y 轴刻度精度
# plt.yticks(np.arange(0, 100, 5))

plt.savefig('../../imgs/所有 OpenID 的学员使用数量4-6.png', dpi=800, bbox_inches='tight', pad_inches=1)
plt.show()
```
